```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#generate normally distributed data 
# setting mean std and amount of points 
data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
# make histogram with parapmeters 
#fairly normally distirbuted set not perfect wont be complelty perfect have outlier son each side not much sqew 
# use hist to see the distribution of data and know what need to do 
#analyse the type of distibution you have and select the mjodel you would fit
# semi colon at end would hide the count 
plt.hist(data, bins=50, alpha=0.7, color='blue')
```

```python
#QQ plot helps generate thorietical quantiles of what thoeretical perfect normal would exist and compare to your own 
# on x axis its if you had perrfect on y its the values you actuially have to be perfectly allinged it would all have to be perfectly on the red line 
#identitifeies if data visually if it is close to normal 

stats.probplot(data, dist='norm', plot=plt);
```

poisson distribution 
- for discrete values normal would be more continuous variable 
- discrete variable would be like coutns (ie number of people in store today)
- when have count and business problem dont want normal distribution because cant have 12.5 people more appropriate would be poisson 
distribution belwo would be like most transactions happen between 8 and 10 is where we see most of the time 
- bins will be continuous but data will be discrete 

```python
from scipy.stats import poisson
#gnerate data 
data = poisson.rvs(mu=10, size=1000)
#create histogram 
plt.hist(data, bins=30, alpha=0.7, color='green');
```

statisitcal test depend on type of distribution 
parametric test--> normal  (t test)
other --. non parametric ( test of proportion/ crisco wallace test)
#number of exposures to buy something 
#could model how many impression must be given --> discrete 

```python
# t test 
#simulated database 
# if we have big enough sample size should see its stat sig 
# good idea to assum variances are not equal 
# two sided greater or less 
store_a_sales = np.random.normal(30000, 5000, 100)
store_b_sales = np.random.normal(35000, 5000, 100)
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(store_a_sales, store_b_sales, equal_var=False, alternative='two-sided')
print(p_value) #example of t test check the defult assumptions so you know what it is assuming 
```

mann whitney test 
- robust for non normal data 
-less precise than t test for normal data 
can make business decision on non parametric 

```python
from scipy.stats import mannwhitneyu
u_stat, p_value = mannwhitneyu(store_a_sales, store_b_sales)
print(p_value)
```

```python
cust_level = df.groupby('Customer_ID')[['Customer_Age_Months', 'Gender', 'Campaign_Exposure',
       'Transaction_Category', 'Purchase_Amount', 'Premium_Product']].agg({'Customer_Age_Months':'max', 'Campaign_Exposure':'max', 
                                                                          'Purchase_Amount':'mean', 'Premium_Product':'mean'})
campaign_cx = cust_level[cust_level['Campaign_Exposure']==1]
reg_cx = cust_level[cust_level['Campaign_Exposure']==0]
sns.boxplot(x = 'Campaign_Exposure', y = 'Purchase_Amount', data=cust_level)
sns.boxplot(x = 'Campaign_Exposure', y = 'Premium_Product', data=cust_level)
```

```python
#ttest first 
# ttest first
t_stat, p_value = ttest_ind(campaign_cx['Purchase_Amount'], reg_cx['Purchase_Amount'])
print(p_value)
```

```python
# is alternative greater significant 
t_stat, p_value = ttest_ind(campaign_cx['Purchase_Amount'], reg_cx['Purchase_Amount'], alternative='greater')
print(p_value)
```
