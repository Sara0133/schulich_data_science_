**STEP 3: Data Engineering**
- Cleaning and Processing 

```python
#Turn Date_of_churn & Subscription_start_date into Datetime data 
df['Date_of_Churn'] = pd.to_datetime(df['Date_of_Churn'])
df['Subscription_Start_Date'] = pd.to_datetime(df['Subscription_Start_Date'])
```

```python
#Look at data type to ensure they have been converted 
df.info()
```

```python
#Add Sub_length column: this column will display the time with the subscription before churn 
df['sub_length'] = (df['Date_of_Churn'] - df['Subscription_Start_Date']).dt.days
```

```python
#look at data to ensure it has been added 
df.head(25)
```

**STEP 4: Explore and understand the factors contributing to churn**
- To find the relevant factors for churn that have an influence I will run visualization of differnt factors against churn and look at rate of churn between different values within each column
    

```python
# first look at Heatmap: this will show the linear correlation of different numerical values 
numeric_columns = df.select_dtypes(include=['float64', 'int64', 'datetime64[ns]']).columns
corr_matrix = df[numeric_columns].corr()
plt.figure(figsize = (20,20))
sns.heatmap(df[numeric_columns].corr(), annot=True)
```

<u>From this I conclude</u>
- Rating given has high  negative corelation to churn
- Monthly hours watched, subscription start date and age are somewhat negatively correlated to churn 


Now I will explore further each of the significant correlations to churn


```python
#Monthly Hours Watched 
#boxplot of monthy hours watched against churn 
# this will show Distribution of monthly hours watched for churned (1) and non-churned (0) users (median, quartiles, and potential outliers)
sns.boxplot(x='Churned', y='Monthly_Hours_Watched', data=df)
plt.title('Box Plot of Monthly Hours Watched vs Churn')
plt.xlabel('Churned')
plt.ylabel('Monthly Hours Watched')

```

<u>From this I can conclude </u>
- Churned users watch somewhat fewer hours than non churned 

```python
#Age
#boxplot of Age against churn 
# this will show Distribution of Age for churned (1) and non-churned (0) users (median, quartiles, and potential outliers)
sns.boxplot(x='Churned', y='Age', data=df)
plt.title('Box Plot of Age vs Churn')
plt.xlabel('Churned')
plt.ylabel('Age')


# I will also make a box plot with age groups to see if certain age ranges have a greater effect 
#Create age groups
bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 70]
labels = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-70']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

#Box plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Age_Group', y='Churned')
plt.title('Bar Plot of Age group vs Churn')
plt.xlabel('Age')
plt.ylabel('Churn')

#countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_Group', hue='Churned', data=df)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group')

```

From this I can conclude 
- membership skews towards an older age but our biggest churneres are younger 

```python
# Rating given 
#boxplot of rating given against churn 
# this will show Distribution of Rating Given for churned (1) and non-churned (0) users (median, quartiles, and potential outliers)
sns.boxplot(x='Churned', y='Rating_Given', data=df)
plt.title('Box Plot of  Rating_Given vs Churn')
plt.xlabel('Churned')
plt.ylabel('Rating Given')

# hisplot to see the distribution of ratings given for churned and non churned users
plt.figure(figsize=(10, 6))
sns.countplot(x='Rating_Given', hue='Churned', data=df)
plt.title(' Ratings Given for Churned and Not Churned ')
plt.xlabel('Rating Given')
plt.ylabel('Count')

#plot for the percentage of people who churned at each rating
# percentage of people who churned at each rating
rating_churn_data = df.groupby('Rating_Given')['Churned'].value_counts(normalize=True).unstack() * 100
#NaN values fill with 0 
rating_churn_data = rating_churn_data.fillna(0)
# New columns 
rating_churn_data.columns = ['Not Churned (%)', 'Churned (%)']

plt.figure(figsize=(10, 6))
sns.barplot(x=rating_churn_data.index, y=rating_churn_data['Churned (%)'])
plt.title('Percentage of People Who Churned at Each Rating')
plt.xlabel('Rating Given')
plt.ylabel('Churned (%)')


```

<u>From this I can conclude that the impact of monthly hours watched is small </u>
- Median monthly hours watched is higher for non-churned users compared to churned users
- Interquartile range  wider for non churned showing more variability in the monthly hours watched
- Some outliers in the non-churned group with very high monthly hours watched
- A few outliers in churned with high hours watched 

I will Now look at Subscription Length column I created to see its impact on churn 

```python
#boxplot of Subscription Length against churn 
# this will show Distribution of Subscription Length for churned (1) and non-churned (0) users (median, quartiles, and potential outliers)
#create column for churned users 
#Analyze the subscription lengths of churned users
churned_users = df[df['Churned'] == 1]

# Calculate statistics for subscription lengths
subscription_length_stats = churned_users['sub_length'].describe()

# Group the subscription lengths into bins
bins = [0, 30, 90, 180, 365, 730, 1095, float('inf')]
labels = ['<1 month', '1-3 months', '3-6 months', '6-12 months', '1-2 years', '2-3 years', '>3 years']
churned_users['Subscription_Length_Group'] = pd.cut(churned_users['sub_length'], bins=bins, labels=labels)

# Count the number of churned users in each subscription length group
subscription_length_group_counts = churned_users['Subscription_Length_Group'].value_counts().sort_index()

# Plot the distribution of subscription length groups
plt.figure(figsize=(10, 6))
subscription_length_group_counts.plot(kind='bar')
plt.title('Distribution of Subscription Lengths Among Churned Users')
plt.xlabel('Subscription Length Group')
plt.ylabel('Number of Churned Users')
plt.xticks(rotation=45)
plt.show()
```

<u>From this I can conclude </u>
- churned users do not have the subscription for more than one year 
- many churned users have a subscription for about 6 months to a year
