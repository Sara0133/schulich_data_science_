```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/delinai/schulich_ds1_2024/main/Datasets/Assignment1_StreamFlow_Subscription_Data.csv')
```

```python
# Descriptive statistics
data.describe()
```

```python
data.describe()
```

```python
#check for missing values
data.isnull().sum()
```

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data[numeric_columns].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Numerical Data Correlations')
```

```python
#from heatmap can see the only significany linear correlation of numerical data is reason given to churn
```

```python
# Calculate the percentage of churn by reason given
churn_counts = data['Churn_Reason'].value_counts(normalize=True) * 100
```

```python
# Churn distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Churned', data=data)
plt.title('Churn Distribution')
plt.show()
```

```python
# Subscription tier distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Subscription_Type', data=data)
plt.title('Subscription Type Distribution')
plt.show()
```

```python
# Churn by subscription tier
plt.figure(figsize=(10, 8))
sns.countplot(x='Subscription_Type', hue='Churned', data=data)
plt.title('Churn by Subscription Type')
plt.show()
```
