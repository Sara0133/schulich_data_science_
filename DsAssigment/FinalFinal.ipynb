STEP 3: Data Engineering 
- Cleaning and Processing 

```python
#turn Date_of_churn & Subscription_start_date into Datetime data 
df['Date_of_Churn'] = pd.to_datetime(df['Date_of_Churn'])
df['Subscription_Start_Date'] = pd.to_datetime(df['Subscription_Start_Date'])
```

```python
#look at data type to ensure they have been converted 
df.info()
```

```python
#Add Sub_length column: this column will display the time with the subscription before churn 
df['sub_length'] = (df['Date_of_Churn'] - df['Subscription_Start_Date']).dt.days
```

```python
#look at data t to ensure it has been added 
df.head(25)
```

```python
# first look at Heatmap: this will show the linear corealtion of different numerical values 
numeric_columns = df.select_dtypes(include=['float64', 'int64', 'datetime64[ns]']).columns
corr_matrix = df[numeric_columns].corr()
plt.figure(figsize = (20,20))
sns.heatmap(df[numeric_columns].corr(), annot=True)
```

From this I conclude
- Rating given has high  negative corelation to churn
- Montly hours watched, subscription start date and age are somewhat negatively corealted to churn

Now I will explore further each of the significant corelations to churn


```python
#Monthly Hours Watched 
#boxplot of monthy hours watched against churn 
# this will show Distribution of monthly hours watched for churned (1) and non-churned (0) users (median, quartiles, and potential outliers)
sns.boxplot(x='Churned', y='Monthly_Hours_Watched', data=df)
plt.title('Box Plot of Monthly Hours Watched vs Churn')
plt.xlabel('Churned')
plt.ylabel('Monthly Hours Watched')

```

```python
#Age
#boxplot of Age against churn 
# this will show Distribution of Age for churned (1) and non-churned (0) users (median, quartiles, and potential outliers)
sns.boxplot(x='Churned', y='Age', data=df)
plt.title('Box Plot of Age vs Churn')
plt.xlabel('Churned')
plt.ylabel('Age')


# I will also Make a box plot with age groups to see if certain age ranges have a greater effect 
#Create age groups
bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 70]
labels = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-70']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

#Box plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Age_Group', y='Churned')
plt.title('Bar Plot of Age group vs Churn')
plt.xlabel('Age')
plt.ylabel('Churn')

#countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_Group', hue='Churned', data=df)
plt.xlabel('Age_Group')
plt.ylabel('Count')
plt.title('Age Group')

```

From this I can conclude 
- membership skews towards an older age but our biggest churneres are younger 

```python
# Rating given 
#boxplot of Rating given against churn 
# this will show Distribution of Rating Givin for churned (1) and non-churned (0) users (median, quartiles, and potential outliers)
sns.boxplot(x='Churned', y='Rating_Given', data=df)
plt.title('Box Plot of  Rating_Given vs Churn')
plt.xlabel('Churned')
plt.ylabel('Rating_Given')

# hisplot to see the distribution of ratings given for churned and not churned users
plt.figure(figsize=(10, 6))
sns.countplot(x='Rating_Given', hue='Churned', data=df)
plt.title(' Ratings Given for Churned and Not Churned ')
plt.xlabel('Rating Given')
plt.ylabel('Count')

#plot for the ercentage of people who chured at each rating
# percentage of people who churned at each rating
rating_churn_data = df.groupby('Rating_Given')['Churned'].value_counts(normalize=True).unstack() * 100
#NaN values fill with 0 
rating_churn_data = rating_churn_data.fillna(0)
# New columns 
rating_churn_data.columns = ['Not Churned (%)', 'Churned (%)']

plt.figure(figsize=(10, 6))
sns.barplot(x=rating_churn_data.index, y=rating_churn_data['Churned (%)'])
plt.title('Percentage of People Who Churned at Each Rating')
plt.xlabel('Rating Given')
plt.ylabel('Churned (%)')


```

```python
#Reason_for_Churn
# I will use a pie chart to see how the different reasons for churn contribute to the total churn value 

#Include only churned users
churned_users = df[df['Churned'] == 1]
#Percentage of each reason for churn, excluding NaN values
Reason_Count = churned_users['Reason_for_Churn'].dropna().value_counts()
Reason_Percentages = (Reason_Count / Reason_Count.sum()) * 100
# Create a pie
plt.figure(figsize=(10, 6))
plt.pie(Reason_Percentages, labels=Reason_Percentages.index, autopct='%1.1f%%')
plt.title('Percentage of Churn by Reason for Churn')

```
